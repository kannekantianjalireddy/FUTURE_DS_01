```python
# General KPIs
```

```python
print("===== GENERAL KPIs =====")
print("Total Products:", df.shape[0])
print("Total Columns:", df.shape[1])
print("Total Missing Values:", df.isnull().sum().sum())
```

```python
# Pricing KPIs
```

```python
print("\n===== PRICING KPIs =====")
print("Average Price:", df['price'].mean())
print("Minimum Price:", df['price'].min())
print("Maximum Price:", df['price'].max())
print("\nAverage MRP:", df['mrp'].mean())
print("Minimum MRP:", df['mrp'].min())
print("Maximum MRP:", df['mrp'].max())
```

```python
# Margin KPIs
```

```python
print("\n===== MARGIN KPIs =====")
print("Average Margin %:", df['margin_percentage'].mean())
print("Minimum Margin %:", df['margin_percentage'].min())
print("Maximum Margin %:", df['margin_percentage'].max())
```

```python
# Inventory KPIs
```

```python
print("\n===== INVENTORY KPIs =====")
print("Average Shelf Life:", df['shelf_life_days'].mean())
print("Average Min Stock Level:", df['min_stock_level'].mean())
print("Average Max Stock Level:", df['max_stock_level'].mean())
```

```python
# Category-wise KPI Analysis
```

```python
category_kpi = df.groupby('category').agg({
    'price': 'mean',
    'mrp': 'mean',
    'margin_percentage': 'mean',
    'product_id': 'count'
}).rename(columns={'product_id': 'Total Products'})
print("\n===== CATEGORY-WISE KPIs =====")
print(category_kpi)
```

```python
plt.figure()
plt.hist(df['price'], bins=20)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('category')['price'].mean().plot(kind='bar')
plt.title("Average Price by Category")
plt.xlabel("Category")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

```python
# Insight Generation
```

```python
df['estimated_profit'] = df['price'] * df['margin_percentage'] / 100
```

```python
# Top 10 Most Expensive Products
```

```python
top_expensive = df.sort_values(by='price', ascending=False).head(10)
top_expensive[['product_name', 'category', 'price']]
```

```python
#Top 10 Most Profitable Products
```

```python
top_profit = df.sort_values(by='estimated_profit', ascending=False).head(10)
top_profit[['product_name', 'category', 'estimated_profit']]
```

```python
category_analysis = df.groupby('category').agg({
    'price': 'mean',
    'margin_percentage': 'mean',
    'estimated_profit': 'sum',
    'product_id': 'count'
}).rename(columns={'product_id': 'total_products'})

category_analysis.sort_values(by='estimated_profit', ascending=False)
```

```python
low_margin = df[df['margin_percentage'] < 20]
low_margin[['product_name', 'category', 'margin_percentage']].head()
```

```python
overstock = df[df['max_stock_level'] > 90]
overstock[['product_name', 'category', 'max_stock_level']]
```

```python
low_shelf_life = df[df['shelf_life_days'] < 10]
low_shelf_life[['product_name', 'category', 'shelf_life_days']]
```

#Relationship Insights

```python
correlation = df[['price', 'mrp', 'margin_percentage', 
                  'shelf_life_days', 'min_stock_level', 
                  'max_stock_level', 'estimated_profit']].corr()
correlation
```

#Visualization for Insight

```python
df.groupby('category')['estimated_profit'].sum().plot(kind='bar')
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.show()
```
