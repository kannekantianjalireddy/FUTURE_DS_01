```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv("blinkit_products.csv")
```

```python
# show basic information
print("Dataset Shape:", df.shape)
print(df.head())
```

```python
# create discount percentage column
df["discount_percent"] = ((df["mrp"] - df["price"]) / df["mrp"]) * 100
```

```python
# basic statistics
print("\nAverage Price:", df["price"].mean())
print("Average MRP:", df["mrp"].mean())
print("Average Discount %:", df["discount_percent"].mean())
```

```python
# category performance
category_count = df["category"].value_counts()
print("\nTop Categories:\n", category_count)
```

```python
# average price per category
category_price = df.groupby("category")["price"].mean()
print("\nAverage Price by Category:\n", category_price)
```

```python
# stock analysis
print("\nAverage Minimum Stock:", df["min_stock_level"].mean())
print("Average Maximum Stock:", df["max_stock_level"].mean())
```

```python
# shelf life analysis
print("\nAverage Shelf Life:", df["shelf_life_days"].mean())
```

```python
# category distribution
plt.figure()
category_count.head(10).plot(kind="bar")
plt.title("Top Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45)
plt.show()
```

```python
# price vs mrp
plt.figure()
plt.scatter(df["mrp"], df["price"])
plt.title("MRP vs Selling Price")
plt.xlabel("MRP")
plt.ylabel("Price")
plt.show()
```

```python
# discount distribution
plt.figure()
sns.histplot(df["discount_percent"], bins=20)
plt.title("Discount Percentage Distribution")
plt.xlabel("Discount %")
plt.ylabel("Count")
plt.show()
```

```python
# shelf life distribution
plt.figure()
sns.histplot(df["shelf_life_days"], bins=20)
plt.title("Shelf Life Distribution")
plt.xlabel("Shelf Life Days")
plt.ylabel("Count")
plt.show()
```

```python
# stock levels
plt.figure()
plt.scatter(df["min_stock_level"], df["max_stock_level"])
plt.title("Stock Level Analysis")
plt.xlabel("Min Stock")
plt.ylabel("Max Stock")
plt.show()
```
