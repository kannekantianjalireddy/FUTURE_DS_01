```python
import pandas as pd
df = pd.read_csv("blinkit_products.csv")
avg_margin = df["margin_percentage"].mean()
print("Average Margin:", avg_margin)
```

```python
short_life = df[df["shelf_life_days"] <= 3]
print(short_life[["product_name","category","shelf_life_days"]])
```

```python
low_stock = df[df["min_stock_level"] > df["max_stock_level"] * 0.7]
print(low_stock[["product_name","min_stock_level","max_stock_level"]])
```

```python
df["discount"] = df["mrp"] - df["price"]
top_discount = df.sort_values("discount", ascending=False).head(10)
print(top_discount[["product_name","price","mrp","discount"]])
```

```python
avg_margin = df["margin_percentage"].mean()
print("\nAverage Margin:", avg_margin)
```

```python
top_margin = df.sort_values(by="margin_percentage", ascending=False).head(10)
print("\nTop Margin Products:")
print(top_margin[["product_name","margin_percentage"]])
```
